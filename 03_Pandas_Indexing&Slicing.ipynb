```python
import pandas as pd
import numpy as np
```

```python
d = pd.date_range('20191101', periods = 10, freq='Y')
```

```python
r = np.random.randn(10,4)
df= pd.DataFrame(r, index= d, columns=list('ABCD'))
print(df)
```

```python
print(df['A'])
```

```python
print(df[['A','B']])
```

```python
df[0:3]
```

```python
df.iloc[:,-1:-2:-1]
```

```python
df.iloc[0:3,-1:-3:-1]
```

```python
df
```

```python
df.iloc[:7:2,:-4:-2]
```

```python
df.iloc[::,:-4:-1]
```

```python
df.iloc[::-3,-2:-5:-2]
    
```

```python
df.iloc[::-3,-2::-2]
```

```python
df.iloc[::-3,-1:-5:-3]
```

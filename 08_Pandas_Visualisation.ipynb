```python
import pandas as pd
import numpy as np
import os
```

```python
d = pd.date_range('20191101', periods = 10, freq='Y')
r = np.random.randn(10,4)
df= pd.DataFrame(r, index= d, columns=list('ABCD'))
print(df)
```

```python
df['A'].plot.box()
```

```python
df['A'].plot.box(vert=False)
```

```python
df['A'].plot()
```

```python
df['A'].plot.bar()
```

```python
df['A'].plot.barh()
```

```python
df['A'].plot.hist()
```

```python
df['A'].plot.hist(bins=3)
```

```python
df['A'].plot.area()
```

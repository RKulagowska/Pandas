```python
import pandas as pd
import numpy as np
```

```python
students= pd.DataFrame({'s_name':['Jan','Otto', 'Tom', 'John'],'city':['Kraków','Łódź','Lwów','Tarnów'],'contact':[999888777,876543678,876434789,965234789]})
course= pd.DataFrame([['Jan', 'Python',100],['Otto','ML',140],['Tom','R',200],['John','SQL',205]], columns= ['s_name','course','fee'])
print(students)
print(course)
```

```python
pd.merge(students,course,on='s_name')
```

```python
pd.merge(students,course,on='s_name', how = 'inner')
```

```python
students=students.rename(columns={'s_name':'student_name'})
print(students)
```

```python
students['student_name'][1]='Lama'
print(students)
```

```python
students.loc[3,'student_name']='Josh'
print(students)
```

```python
students.iloc[2,0]='Jack'
print(students)
```

```python
pd.merge(students,course,left_on='student_name', right_on='s_name', how = 'inner')
```

```python
pd.merge(students,course,left_on='student_name', right_on='s_name', how = 'outer', sort=True)
```

```python
pd.merge(students,course,left_on='student_name', right_on='s_name', how = 'left')
```

```python
pd.merge(students,course,left_on='student_name', right_on='s_name', how = 'right')
```

```python
pd.merge(students,course,left_on='student_name', right_on='s_name', how = 'right', sort=True)
```
